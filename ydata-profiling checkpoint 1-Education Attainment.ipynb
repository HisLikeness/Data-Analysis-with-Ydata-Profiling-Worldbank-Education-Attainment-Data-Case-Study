```python
# Look for missing values in the dataset. You can use the report generated by ydata to identify missing values.
Missing_values = World_Education_Attainment_and_Enrolment.isnull().sum()
print(Missing_values)
```

```python
# Fill missing values with the mean
# Select numeric columns
Numeric_columns = World_Education_Attainment_and_Enrolment.select_dtypes(include=["int64", "float64"]).columns

World_Education_Attainment_and_Enrolment[Numeric_columns] = World_Education_Attainment_and_Enrolment[Numeric_columns].fillna(World_Education_Attainment_and_Enrolment[Numeric_columns].mean())

# confirming that there are no more missing values:
World_Education_Attainment_and_Enrolment.isnull().sum()
```

```python
# Determining duplicates cell

# Checking for duplicates
duplicates = World_Education_Attainment_and_Enrolment.duplicated().sum()
print(f"The number of duplicate rows are: {duplicates}")

# Display duplicate rows
duplicate_rows = World_Education_Attainment_and_Enrolment[World_Education_Attainment_and_Enrolment.duplicated(keep=False)]
print(duplicate_rows)

# Drop duplicates
World_Education_Attainment_and_Enrolment.drop_duplicates(inplace=True)
```

```python
# Determining the level of correlations between the variables using heatmap
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation matrix
corr_matrix = World_Education_Attainment_and_Enrolment[Numeric_columns].corr()

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, 
            annot=True, 
            cmap='coolwarm', 
            square=True)

# Set title and show plot
plt.title('Variables Correlation Heatmap')
plt.show()

# Save to file
plt.savefig('correlation_heatmap.png')
```
